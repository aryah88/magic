```python
from pathlib import Path
import cv2
import os

def create_folder_structure(base_path):
    for split in ['train', 'test']:
        for label in ['Normal', 'Violence', 'Weaponized']:
            path = Path(base_path) / split / label
            path.mkdir(parents=True, exist_ok=True)
    print(f"✅ Folder structure created under: {base_path}")

def extract_one_frame(video_path, save_path, label, index):
    cap = cv2.VideoCapture(str(video_path))
    if not cap.isOpened():
        print(f"❌ Cannot open video: {video_path}")
        return
    ret, frame = cap.read()
    if ret:
        label_folder = Path(save_path) / label
        label_folder.mkdir(parents=True, exist_ok=True)
        filename = label_folder / f"{label}_{index}.jpg"
        cv2.imwrite(str(filename), frame)
        print(f"✅ Saved: {filename}")
    else:
        print(f"⚠️ Failed to read: {video_path}")
    cap.release()

def process_videos(source_dir, target_dir, split):
    print(f"🔍 Processing {split.upper()} videos...")
    for label in ['Normal', 'Violence', 'Weaponized']:
        class_dir = Path(source_dir) / split / label
        video_paths = list(class_dir.glob("*.avi"))
        print(f"📂 {label}: Found {len(video_paths)} videos in {class_dir}")

        for idx, video_path in enumerate(video_paths):
            extract_one_frame(video_path, Path(target_dir) / split, label, idx)

# -----------------------------
# CONFIG
source_dataset_dir = Path("archive/SCVD/SCVD_converted_sec_split")
target_frames_dir = Path("scvd_frames")
# -----------------------------

create_folder_structure(target_frames_dir)
process_videos(source_dataset_dir, target_frames_dir, 'train')
process_videos(source_dataset_dir, target_frames_dir, 'test')
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.models import Model
from tensorflow.keras.layers import GlobalAveragePooling2D, Dense
from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt

IMG_SIZE = 224
BATCH_SIZE = 16
EPOCHS = 10
```

```python
train_dir = "scvd_frames/train"
test_dir = "scvd_frames/test"

datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2,
    horizontal_flip=True,
    rotation_range=10,
    zoom_range=0.1
)

train_generator = datagen.flow_from_directory(
    train_dir,
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='training'
)

val_generator = datagen.flow_from_directory(
    train_dir,
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='validation'
)
```

```python
base_model = MobileNetV2(include_top=False, input_shape=(IMG_SIZE, IMG_SIZE, 3), weights='imagenet')

x = GlobalAveragePooling2D()(base_model.output)
output = Dense(3, activation='softmax')(x)

model = Model(inputs=base_model.input, outputs=output)

model.compile(optimizer=Adam(learning_rate=1e-4),
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
history = model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=EPOCHS,
    verbose=1
)
```

```python

model.save("scvd_violence_model.h5")
print("✅ Model saved as scvd_violence_model.h5")
```

```python
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Val Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.title('Training & Validation Accuracy')
plt.grid(True)
plt.show()
```

```python
from tensorflow.keras.models import load_model
import cv2
import numpy as np

model = load_model("scvd_violence_model.h5")
IMG_SIZE = 224
CLASSES = ['Normal', 'Violence', 'Weaponized']
```

```python
def predict_frame(frame):
    img = cv2.resize(frame, (IMG_SIZE, IMG_SIZE))
    img = img.astype("float32") / 255.0
    img = np.expand_dims(img, axis=0)
    pred = model.predict(img)[0]
    label = CLASSES[np.argmax(pred)]
    confidence = np.max(pred)
    return label, confidence
```

```python
cap = cv2.VideoCapture(0)  # Use 0 for webcam OR replace with path to video

frame_count = 0
fps_interval = 30  # Check 1 frame per ~1 sec (if video is 30 FPS)

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Only predict every nth frame
    if frame_count % fps_interval == 0:
        label, conf = predict_frame(frame)
    frame_count += 1

    # Display prediction on the frame
    text = f"{label} ({conf*100:.1f}%)"
    cv2.putText(frame, text, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 
                1, (0, 255, 0) if label == 'Normal' else (0, 0, 255), 2)

    cv2.imshow("Violence Detection", frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
